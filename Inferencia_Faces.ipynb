```python
from ultralytics import YOLO
import cv2
import mediapipe as mp
```

```python
# Carregar o modelo de classificação treinado
#Nesse caminho você deve colocar o caminho do seu modelo que foi criado com img 'boas' e 'ruins'
model = YOLO("/content/drive/MyDrive/classification_faces/modelo/faces_best.pt")
```

```python
# Inicializar o MediaPipe Face Detection
# Mediapipe é uma biblioteca de aprendizado de máquina desenvolvida pelo Google, que fornece soluções prontas para tarefas comuns de visão computacional.
#Nesse caso, estamos usando a detecção de rosto.
# Vai precisar instalar a lib mediapiape, caso não tenha instalado.
# pip install mediapipe
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
```

```python
# Caminho da imagem
# Nesse caso a classificação será feita em apenas uma imagem.
# Para fazer a classificação em várias imagens, você pode usar um loop para iterar sobre as imagens e salvar em pastas boas e ruins
image_path = "/content/drive/MyDrive/classification_faces/img_teste/798ac82f-5a5a-4ec7-9798-f9cbfd6f3d4e.jpeg"
# Faz a leitura da imagem usando a função imread do OpenCV
image = cv2.imread(image_path)
# Converte a imagem de BGR para RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Inicializa a detecção de rosto
# O modelo de detecção de rosto é inicializado com a seleção do modelo 0 e uma confiança mínima de detecção de 0.5.
# O modelo_selection=0 refere-se ao modelo de detecção de rosto padrão do MediaPipe.
# O min_detection_confidence=0.5 define a confiança mínima para considerar uma detecção válida.
# O MediaPipe fornece dois modelos de detecção de rosto: o modelo 0 é mais rápido, mas menos preciso, enquanto o modelo 1 é mais lento, mas mais preciso.
with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
    results = face_detection.process(image_rgb)
    # Condicional para verificar se há detecções de rosto
    # A variável results contém as detecções de rosto feitas pelo modelo.
    # A propriedade detections contém uma lista de detecções, onde cada detecção é um objeto que contém informações sobre a localização do rosto.
    if results.detections:
        for detection in results.detections:
            bbox = detection.location_data.relative_bounding_box
            h, w, _ = image.shape
            x = int(bbox.xmin * w)
            y = int(bbox.ymin * h)
            width = int(bbox.width * w)
            height = int(bbox.height * h)

            x, y = max(0, x), max(0, y)
            x2, y2 = min(w, x + width), min(h, y + height)

            # Recorta o rosto da imagem original
            # A variável face_crop contém a imagem do rosto recortada da imagem original.
            # A função max(0, x) garante que as coordenadas não sejam negativas.
            # A função min(w, x + width) garante que as coordenadas não ultrapassem os limites da imagem.
            # A variável face_crop contém a imagem do rosto recortada da imagem original.
            face_crop = image[y:y2, x:x2]
            # Verifica se o rosto foi detectado
            if face_crop.size == 0:
                continue

            # Inferência com modelo YOLO de classificação
            # A função predict() do modelo YOLO é chamada para fazer a classificação da imagem do rosto.
            # A variável results_cls contém os resultados da classificação.
            # A função predict() recebe a imagem do rosto, o tamanho da imagem (imgsz) e a confiança mínima (conf).
            results_cls = model.predict(source=face_crop, imgsz=224, conf=0.5, verbose=False)
            # Verifica se há resultados de classificação
            # A variável results_cls contém os resultados da classificação.
            # A propriedade names contém os nomes das classes e a propriedade probs contém as probabilidades de cada classe.
            # A propriedade top1 contém o índice da classe com a maior probabilidade.
            label = results_cls[0].names[int(results_cls[0].probs.top1)]
            confidence = results_cls[0].probs.data[int(results_cls[0].probs.top1)]

            # Desenhar resultado
            cv2.rectangle(image, (x, y), (x2, y2), (0, 255, 0), 1)
            cv2.putText(image, f"{label} ({confidence:.2f})", (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 1)
    else:
        print("Nenhum rosto detectado.")
```
